```python
import redis
```

```python
try:
    r = redis.Redis(host='localhost', port=6379, db=0)
    response = r.ping()
    if response:
        print("Redis connection successful!")
    else:
        print("Redis server did not respond to PING.")
except redis.exceptions.ConnectionError as e:
    print(f"Could not connect to Redis: {e}")
```
